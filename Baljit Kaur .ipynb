```python
gun_deaths['date'] = gun_deaths['month'].map(str) + '-' + gun_deaths['year'].map(str)
gun_deaths1 = gun_deaths.set_index('date')
gun_deaths1.head()
```

# Analysis with plots of important columns

```python
plot = gun_deaths1['sex'].value_counts().plot(kind='bar',figsize=(14,8), title="Bar plot for both gender")
plt.xlabel("gender")
plt.ylabel("number of males and females")
```

From above plot, it is clear that the number of women is very less as compare to men in this shooting who were died.

```python
plot = gun_deaths1['place'].value_counts().plot(kind='bar',figsize=(14,8), title="number of intent at different places",)
plt.xlabel("different places")
plt.ylabel("total number of intent")
```

This plot shows that number of intent at home is very high than other places which is almost 60000. Moreover, it is almost null at construction, residential institution and sports areas.

```python
sns.pairplot(gun_deaths1)
```

```python
sns.countplot(x="education",hue="sex", data=gun_deaths1)
plt.xlabel("Education")
plt.ylabel("number of males and females")
```

This plot depicts that greater number of both males and females were qualified in General Equivalency Diploma or High School Diploma which is above from 35000 for men and more than 5000 for women. BA+ was less intrested among both genders than other fields of education.
